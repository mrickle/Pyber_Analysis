```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
get_data = pd.read_csv('resources/got.csv')
get_data
```

```python
# Get attacker and defender king data
attacker = get_data['attacker_king'].value_counts()
attacker
```

```python
defender = get_data['defender_king'].value_counts()
defender
```

```python
# Get total battle data
#battle_data = attacker_data.add(defender_data, fill_value=0)
got_df = attacker.add(defender, fill_value=0)
got_df
```

```python
# Configure plot and ticks
got_df.plot(kind = 'bar', facecolor = 'green')

# Set textual properties
plt.title('battle data')
#plt.title("Battle Data")
#plt.ylabel("Battles")
#plt.xlabel("king")
#plt.xticks(rotation=45)
#plt.ylabel("Number of Battles Participated In")
#plt.xlabel("King")
# Show plot
plt.show()

# Resize plot to display labels
plt.tight_layout()
```
